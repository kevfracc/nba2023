```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
pd.set_option('display.max_columns', None)
```

```python
#BASE URL
BaseURL = 'https://api-web.nhle.com'
```

```python
#TEAMS
import requests

link = BaseURL+'/v1/standings/now'
f = requests.get(link)

i=0
teamlist=[]
abbrevlist=[]
citylist=[]

while ((len(teamlist)<32)|(len(teamlist)<32)):
    try:
        citylist.append(f.json()['standings'][i]['placeName']['default'])
        teamlist.append(f.json()['standings'][i]['teamCommonName']['default'])
        abbrevlist.append(f.json()['standings'][i]['teamAbbrev']['default'])
        #teamlist=list(set(teamlist))
        i=i+1
    except:
        break
len(teamlist)

Teams=pd.DataFrame()
Teams['City']=citylist
Teams['TeamName']=teamlist
Teams['TeamID']=abbrevlist
Teams
```

```python
#SKATERS
import requests
teamlist=Teams['TeamID']

SeasonPlayerStats=pd.DataFrame()
for team in teamlist:
        
    link = BaseURL + '/v1/club-stats/'+team+'/now'
    f = requests.get(link)

    df=pd.DataFrame(f.json()['skaters'])

    for i in range(0,len(df)):
        firstname=f.json()['skaters'][i]['firstName']['default']
        lastname=f.json()['skaters'][i]['lastName']['default']
        df.iloc[i, df.columns.get_loc('firstName')] = firstname
        df.iloc[i, df.columns.get_loc('lastName')] = lastname
        
    df['Team']=team
    SeasonPlayerStats=pd.concat([SeasonPlayerStats,df])
SeasonPlayerStats

```

```python
link = BaseURL + '/v1/player/'+'8481567'+'/game-log/20232024/2'
f = requests.get(link)
df=pd.DataFrame(f.json()['gameLog'])
df
```

```python
import requests
players=SeasonPlayerStats['playerId']

DailyPlayerStats=pd.DataFrame()
for player in players:
        
    link = BaseURL + '/v1/player/'+str(player)+'/game-log/20232024/2'
    f = requests.get(link)
    df=pd.DataFrame(f.json()['gameLog'])
    df['playerId']=player
    DailyPlayerStats=pd.concat([DailyPlayerStats,df])
#8m57s 1/5/24
#11 min 1/9/24
#9.45 min 1/10/24
```

```python
#DailyPlayerStats
#LastYearStats=DailyPlayerStats[DailyPlayerStats['gameDate']<'2024-01-01']
ThisYearStats=DailyPlayerStats[DailyPlayerStats['gameDate']>='2024-01-01']
ThisYearStats=ThisYearStats.merge(SeasonPlayerStats[['playerId','firstName','lastName']].drop_duplicates(), on='playerId', how='left')

#LastYearStats.to_csv('LastYear2023Stats.csv', index=False)
ThisYearStats.to_csv('ThisYear2024Stats.csv', index=False)
```

```python
#GOALIES
import requests
teamlist=Teams['TeamID']

SeasonGoalieStats=pd.DataFrame()
for team in teamlist:
        
    link = BaseURL + '/v1/club-stats/'+team+'/now'
    f = requests.get(link)

    df=pd.DataFrame(f.json()['goalies'])

    for i in range(0,len(df)):
        firstname=f.json()['goalies'][i]['firstName']['default']
        lastname=f.json()['goalies'][i]['lastName']['default']
        df.iloc[i, df.columns.get_loc('firstName')] = firstname
        df.iloc[i, df.columns.get_loc('lastName')] = lastname
    df['Team']=team
    SeasonGoalieStats=pd.concat([SeasonGoalieStats,df])
SeasonGoalieStats

```

```python
import requests
players=SeasonGoalieStats['playerId']

DailyGoalieStats=pd.DataFrame()
for player in players:
        
    link = BaseURL + '/v1/player/'+str(player)+'/game-log/20232024/2'
    f = requests.get(link)
    df=pd.DataFrame(f.json()['gameLog'])
    df['playerId']=player
    DailyGoalieStats=pd.concat([DailyGoalieStats,df])
#46 sec
```
